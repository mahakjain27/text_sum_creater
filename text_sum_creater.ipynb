```python
from transformers import pipeline

# Corrected model name and task
summarizer = pipeline("summarization", model="google/pegasus-xsum")

def summarize_legal_text(text, max_length=100, min_length=30):
    summary = summarizer(text, max_length=max_length, min_length=min_length, do_sample=False)
    return summary[0]['summary_text']

# Define legal text outside the function
legal_text = """
A legal text refers to a written document that has a legal character and is intended for legal purposes.
It encompasses a wide range of documents, including statutes, contracts, judicial opinions, and legal briefs.
Legal texts are characterized by their formal structure, precise language, and specific legal terminology.
They serve various functions within the legal system, such as conveying legal rights, obligations, and procedures.
"""

# Summarize and print result
summarized_text = summarize_legal_text(legal_text)
print(summarized_text)
```
